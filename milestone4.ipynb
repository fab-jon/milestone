```python
import pandas as pd
import matplotlib.pyplot as plt
datafile = pd.read_csv("clean_school_perception.csv")
```

## Does race play a role in students feeling safe?

```python
race = datafile [["black_safe", "native_american_safe", "asian_safe", "hispanic_safe", "pacific_islander_safe", "white_safe", "multiracial_safe", "other_race_safe"]]
race
```

```python
race.sum()
```

```python
race.sum().plot(kind='bar', color='blue') 
plt.title("Students' Race May Impact Feeling Safe in School")
plt.xlabel("Race")
plt.ylabel("Percent")
plt.show()
```

### In these California school districts, we conclude that race may affect a student feeling safe in school because the percentage for Native American students is about 10% lower than the other students.

## What is the breakdown of safety perception of students who have a low level of connectedness to school? 

```python
connect = datafile [["low_connect_very_safe", "low_connect_safe", "low_connect_neutral", "low_connect_unsafe", "low_connect_very_unsafe"]]
connect
```

```python
connect.sum()
```

```python
connect.sum().plot(kind='barh',color='gray') 
plt.title("Having Low Connection to School Doesn't Impact Feeling of Safety")
plt.xlabel("Percent")
plt.ylabel("Safety Feeling")
plt.show()
```

### According to the bar plot we see that the majority of students who feel a low connection to the school feel "neutral" about their safety

## Does sexual orientation play a role in students feeling very unsafe in school? 

```python
orientation = datafile[["gay_lesbian_very_unsafe", "straight_very_unsafe","unsure_very_unsafe"]]
orientation
```

```python
orientation.sum()
```

```python
orientation.sum().plot(kind='bar', color="purple") 
plt.title("Students Feeling Very Unsafe is Affected by Sexual Orientation")
plt.show()
```

### When it comes to sexual orientation, it is more likely for gay or lesbian students to feel more "very unsafe" compared to students who are straight or unsure of their sexual orientation.

## Does the parents’ level of education affect students who feel neither safe nor unsafe? 

```python
parents = datafile[[ "parents_not_finish_high_school_neutral", "parents_finish_high_school_neutral", "parents_some_college_neutral", "parents_bachelors_neutral"]]
parents
```
